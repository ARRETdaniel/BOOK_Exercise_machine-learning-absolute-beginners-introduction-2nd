```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
import sklearn.externals
import joblib
```

```python
df = pd.read_csv('C:/Users/danie/Documents/CURSOS/courseBook/dataset/Melbourne_housing_FULL.csv')
```

```python
#: Previewing a dataframe
df.head(n=5)
```

```python
# The misspellings of “longitude” and “latitude” are used, as the two misspellings were not corrected in the source file.
del df['Address']
del df['Method']
del df['SellerG']
del df['Date']
del df['Postcode']
del df['Lattitude']
del df['Longtitude']
del df['Regionname']
del df['Propertycount']
```

```python
df.head(n=10)
```

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
df.head(n=10)
```

```python
features_df = pd.get_dummies(df, columns=['Suburb', 'CouncilArea', 'Type'])
```

```python
df.head(n=10)
```

```python
del features_df['Price']
```

```python
X = features_df.to_numpy()
y = df['Price'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = ensemble.GradientBoostingRegressor(
n_estimators=150,
learning_rate=0.1,
max_depth=30,
min_samples_split=4,
min_samples_leaf=6,
max_features=0.6,
loss='huber'
)
```

```python
model.fit(X_train, y_train)
```

```python
joblib.dump(model, 'house_trained_model.pkl')
```

```python
mse = mean_absolute_error(y_train, model.predict(X_train))
print ("Training Set Mean Absolute Error: %.2f" % mse)
```

```python
mse = mean_absolute_error(y_test, model.predict(X_test))
print ("Test Set Mean Absolute Error: %.2f" % mse)
```
